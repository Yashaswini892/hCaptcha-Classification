```python
def data(path):
    lst_of_classes=['airplane', 'bicycle', 'boat', 'motorbus', 'motorcycle', 'seaplane','truck']
    temp= pd.DataFrame(columns = ['path', 'target'])
    for target in lst_of_classes:
        cls_imgs = glob.glob((f"{path}\\{target}\\*.jpg"))
        df = pd.DataFrame({
                'path': cls_imgs,
                'target': [target]*len(cls_imgs),
            })
        temp=pd.concat([temp,df],ignore_index = True)
    return temp
```

```python
#knn
def accuracy_score(ytest,ypred):
    c=0
    for i in range(len(ytest)):
        if ytest[i]==ypred[i]:
            c+=1
    return c/len(ytest)
```

```python
#using distance euclidean distance metic
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=4,p=2)
f=0
for train_idx, test_idx in kfold.split(trdata["ch_nor_gimg"],trdata["label"]):
    test = trdata.loc[test_idx]
    train = trdata.loc[train_idx]
    xtr=[]
    xte=[]
    for i in range(len(train)):
        xtr.append(train["ch_nor_gimg"].iloc[i])
    for i in range(len(test)):
        xte.append(test["ch_nor_gimg"].iloc[i])

    knn.fit(np.array(xtr),train["label"])
    y_pred=knn.predict(np.array(xte))
    ac=accuracy_score(np.array(test["label"]), y_pred)
    print(f"KNN-Classifier Multi-class Set-Accuracy for fold-{f} is {ac*100}")
    f+=1
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import BaggingClassifier, AdaBoostClassifier, RandomForestClassifier
from xgboost import XGBClassifier

Decision_bagg = BaggingClassifier(n_estimators=50, random_state=0, max_features = 5)
svc_bagg = BaggingClassifier(base_estimator=SVC(), n_estimators=10, random_state=0)
Decision_boost = AdaBoostClassifier(n_estimators=200, random_state=0)
rand_forest = RandomForestClassifier(n_estimators=50, random_state=0, max_features = 3)
xgboost=XGBClassifier(n_estimators = 200, random_state=42)
models = {
    'Decison Tree Bagging Classifier': Decision_bagg,
    'SVM Bagging Classifier': svc_bagg,
    'Decision Tree Boosting Classifier': Decision_boost,
    'Random Forest Classifier': rand_forest,
    'xgb':xgboost
}
```

Validation of models based on kfold cross validation.`

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, f1_score
from imblearn.over_sampling import ADASYN
from imblearn.over_sampling import SMOTE

# Stratified K-fold
kfold = StratifiedKFold(shuffle = True)
for name, model in models.items():
    print(f"\nModel - {name}")
    for i, (val_idx, train_idx) in enumerate(kfold.split(xtrain, ytrain)):
        val_x = xtrain.iloc[val_idx].values
        val_y = ytrain.iloc[val_idx].values

        train_x = xtrain.iloc[train_idx].values
        train_y = ytrain.iloc[train_idx].values
        smote= SMOTE(sampling_strategy='minority', random_state=42)
        x_res, y_res = smote.fit_resample(train_x,train_y)


        model.fit(x_res, y_res)
        val_ypr = model.predict(val_x)
        acc = accuracy_score(val_y, np.array(val_ypr))
        f1score=f1_score(val_y, np.array(val_ypr))
        #accS=accuracy_score(np.array(test["ext_label"].values),y_SVC_pred)
        print(f"accuracy of fold {i} is- {acc:0.3f}\nf1 score of fold {i} is {f1score:0.3f}")
```
